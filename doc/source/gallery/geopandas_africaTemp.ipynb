```python
import geopandas
import matplotlib.pyplot as plt
import folium
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

africa = world[world.continent == 'Africa']

africa['area'] = africa.area

africa.explore('area', legend=True)
```

```python
countryList = []

for name in africa.name:
    countryList.append(name)

print(countryList)
```

```python
import requests

temperature = {}

for country in countryList:
    url = 'https://wttr.in/{}?format="%t"'.format(country)
    try:
        res = requests.get(url)
        temperature[country] = res.text
    except:
        pass

print(temperature)
```

```python
africa['temperature'] = africa['name'].map(temperature)

africa
```

```python
africa.explore('area', legend = True, popup = ['name', 'temperature'], style_kwds=dict(color="black"), cmap="Set1", tiles="CartoDB positron", tooltip = False)
```
